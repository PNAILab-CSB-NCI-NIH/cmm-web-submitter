### Load Libraries

```python
import os
import sys
import json

import src.utils as utils

async def main(args):
    out, err = await utils.submit(*args)
    utils.save_json(out, "run.json")
    return out, err
```

### Get directory and paths

```python
# Set input_folder and create dataset info
input_folder = "volumes"
extensions = "pdb,mrc"
data = utils.get_data(input_folder, extensions)
```

### Set variables

```python
# Initialize variables
n_cpus          = 1     # Number of parallel sessions to run
n_files_per_div = 1     # Number of files per parallel session
verbose         = 2     # Verbose level, set it higher to check detailed progress
headless        = True  # 'True': invisible | 'False': visually open browser for inspection
```

```python
# Define parallelism
parallel = True
if parallel:
    n_cpus = 2
    n_files_per_div = 1
```

### Run

```python
# Run
args = (data, verbose, n_cpus, n_files_per_div, headless)
out, err = await main(args)
```
